```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Distribution 1: Beta Distribution

```python
# Distribution Generation & Dataframe
distributionvar1 = np.random.beta(2, 5, 100)
df1 = pd.DataFrame({'Rand':distributionvar1})

# Distribution Histogram
df1.hist()

# Arithmetic Mean & Std. Deviation
print('Arithmetic Mean: {}'.format('%.4f' % np.mean(df1['Rand'])))
print('Standard Deviation: {}'.format('%.4f' % np.std(df1['Rand'])))

# Adding Mean and Std. Deviation Lines on Plot
plt.axvline(np.mean(df1['Rand']), color='b', linestyle='solid', linewidth=2)
plt.axvline(np.mean(df1['Rand']) + np.std(df1['Rand']), color='b', linestyle='dashed', linewidth=2)
plt.axvline(np.mean(df1['Rand']) - np.std(df1['Rand']), color='b', linestyle='dashed', linewidth=2)

# Title of Distribution
plt.title('Beta Distribution')
plt.show()
```

## Additional Assignment

1. Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1.
2. Add them together to create a third variable.
3. Graph the third variable using a histogram.
4. Compute the mean and standard deviation and plot them as vertical lines on the histogram.
5. Evaluate the descriptive statistics against the data.

```python
# Random Size n
n = 100

# Generating Two Normally-Distributed Variables
norm1 = np.random.normal(5, 0.5, n)
norm2 = np.random.normal(10, 1,  n)

# Creating a third variable
norm3 = norm1 + norm2

# Histogram
plt.hist(norm3)

# Mean and Standard Deviation
mean3 = np.mean(norm3)
std3 = np.std(norm3)

#Plotting them as lines on histogram
plt.axvline(mean3, color='r', linestyle='solid', linewidth=2)
plt.axvline(mean3 + std3, color='r', linestyle='dashed', linewidth=2)
plt.axvline(mean3 - std3, color='r', linestyle='dashed', linewidth=2)

plt.show()
```
